## break 與 continue 都是迴圈中斷語法：

```python
i = ['a','b','c','d']
for j in i:
    if j == 'c':
        break
    print(j)     
```

```python
i = ['a','b','c','d']
for j in i:
    if j == 'c':
        continue
    print(j)    
```

# List VS Tuple

+ tuple 與 list 的差異
  + 括號不同：
    + tuple 是用小括號將元素包起來。
    + list 是用中括號將元素包起來。
+ 元素可否變更：
   + tuple 決定了元素的內容就不可以變更。
   + list 可以隨時修改。

## list 介紹
+ list 內資料用逗號隔開，放置於中括號的範圍內。
  + list_name = [1,2,3 ]
+ list 可以儲存不同類型的值或項目，如數字和字串。
  + list_name = [1, 2, ‘Python’, ‘lists’ ] 
+ list 內的資料可以進行增加、更新與移除。
+ 可透過.index( )方法取得指定資料的索引。

```python
list1 = ['H','i']
tuple1 = ('H','i')
print(tuple1)
print(len(tuple1))
print(list1)
print(len(list1))
#tuple1[1] = 'a' #不可修改
list1[1] = 'a'
print(list1)
print(tuple1)
```

### 請問執行後的結果哪一個是對的？

```python
list1 = ['1','2']
list1[1] = '0'
print(list1)
```

## 資料轉換
+ 轉為List 使用 list()函數
+ 轉為tuple 使用 tuple()函數

### 轉為List 

```python
r = range(1,10)
print(r)
print(list(r))
for v in r:
    print(v)
list2 = list("ABCDEF")    
print(list2)
list3 = list((1,2,3,5))
print(list3)
v1,v2,v3,v4 = list3
print(v1,v2,v3,v4)
v1,v2,*v3 = list3
print(v1,v2,v3)
```

### 轉為tuple 

```python
r = range(1,10)
print(type(r))
tuple1 = tuple(r)
print(tuple1)
list1 = [8,9,11,12]
tuple2 = tuple(list1)
print(tuple2)
tuple3 = tuple("ABCDE")
print(tuple3)
v1,v2,*v3 = tuple3
print(v1,v2,v3)
```

## 項目附加資料

```python
list1 = ['a','b','c']
print("附加之前:",list1)
print("附加之前長度:",len(list1))
list1.append(["def","ghij"])
print("附加之後:",list1)
print("附加之後長度=",len(list1))
print(list1[3])
list2 = ['x','y']
list1 += list2
print("附加之後:",list1)
```

## list 項目擴展資料
+ extend 擴展資料：
  + 如果引號括起來的字串有 [ ] 將會拆解成好幾個資料。
  + 可以把其他 list 加入到這個 list 內，擴展多個欄位。

```python
list1 = ['a', 'b', 'c']
print("擴展之前:",list1)
print("擴展之前長度:",len(list1))
list1.extend(["def","ghij"])
print("擴展之後:",list1)
print("擴展之後長度:",len(list1))
print(list1[3])
```
