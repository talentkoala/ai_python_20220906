### 堆疊長條圖

```python
import matplotlib.pyplot as plt
import numpy as np
city = ['Delhi', 'Beijing', 'Washington', 'Tokyo', 'Moscow']
Gender = ['Male', 'Female']
pos = np.arange(len(city))
Happiness_Index_Male = [60, 40, 70, 65, 85]
Happiness_Index_Female = [30, 60, 70, 55, 75]
plt.bar(pos,Happiness_Index_Male,color="blue",edgecolor='black')
plt.bar(pos,Happiness_Index_Female,color="pink",edgecolor="black",bottom =Happiness_Index_Male)
plt.xticks(pos,city)
plt.xlabel('City',fontsize=16)
plt.ylabel("Happiness_Index",fontsize=16)
plt.title("Staked Barchar",fontsize=20)
plt.legend(Gender,loc=2)
plt.show()
```

### Scatter 與 Hist
+ 散佈圖 Scatter：
  + 有助於揭露兩種變數資料的關係  例如國民平均壽命與生育率。
  + 不像折線圖 各點之間不會直線連接。

[Scatter](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html?highlight=scatter#matplotlib.pyplot.scatter)

```python
import matplotlib.pyplot as plt
import numpy as np
h=np.array([159,169,174,165,163,155,180,190])
w=np.array([55,45,70,60,65,45,72,50])
print("身高平均:",np.mean(h))
print("身高中位:",np.median(h))
print("體重平均:",np.mean(w))
print("體重中位:",np.median(w))
plt.scatter(h,w)
plt.show()
```

### Pie 與 Box chart
+ Pie 圓餅圖：顯示變數的分布，例如各區域的銷售量，每一切片大小式依照每個變數的相對強度呈現。
+ Box chart 盒鬚圖：顯示變數分布，方塊顯示中半段數值，兩側的鬚線向兩端延伸：
  + 將多個資料傳遞給某矩陣資料，再使用 boxplot 函數繪製。
  + boxplot( ) 函數的第一個參數為矩陣資料，第二個參數 patch_artist = True 代表填充，第三個參數獲取要繪製的標籤。
  + boxplot( ) 函數接受參 s 數  vert = 0，它繪製水平框圖。

### Pie
+ 派圖  Pie Plots 使用一個完整圓形顯示統計資料。
  + 可增加 explode 製作突增值進行突出來強調顯示
  + 回傳資料中的 patches 代表傳回的色塊物件。
  + 回傳資料中的 texts 代表傳回的各標籤文字的座標與字串。
  + 使用 legend( ) 顯示圖例，第一個參數是色塊物件，第二個是標籤文字，loc 代表位置，best 代表最佳顯示位置。
  + counterclock 是否為逆時針 True逆時鐘

```python
import matplotlib.pyplot as plt
values = [60, 80, 90, 55, 10, 30]
colors = ['b', 'g', 'r', 'c', 'm', 'y']
labels = ['US', 'UK', 'India', 'Germany', 'Australia', 'South Korea']
exploed = (0.1,0,0,0,0,0)
plt.pie(values,colors=colors,labels=values,explode=exploed,counterclock=False,shadow=True)
plt.title("Pie Char!")
plt.legend(labels,loc=3)
plt.show()
```

## Series
類似於一個dictionary 結構包含key與value 
Series的key對應到 index 而value對應到data

```python
import pandas as pd
count = [10,11,12]
s1 = pd.Series(data = count,index = ["A","B","C"])
data2 = {"A":92,"B":83,"C":75}
s2 = pd.Series(data2)
print(s1)
print(s2)
print(data2)
```

### Series修改value

```python
count = [10,11,12]
s1 = pd.Series(data = count,index = ["A","B","C"])
data2 = {"A":92,"B":83,"C":75}
s2 = pd.Series(data2)
print(s1)
tmp = s1.replace(10,200,inplace=False)#inplace False不修改來源
print(s1)
print(tmp)
s2.replace(83,120,inplace=True)
print(s2)
```

## DataFrame
+ DataFrame 是一個二維結構的資料，資料依照行與列的表格方式排列。
+ Column 與 Row 的表達方式如右：
+ Columns 可以是不同類型的資料。
+ 可針對 Columns 與 Rows 進行標記。
+ 可針對 Columns 和 Rows 執行計算。



### DataFrame 的參數
DataFrame 可加入的參數主要有四個：
+ 資料來源：
  + 資料來源，包含 list、ndarray，dictionary、表格、另一個DataFrame、其他 Pandas 物件。
+ Rows 標籤：	
  + 如果沒有設定則會自動進行數值索引顯示。
+ Columns 標籤：	 
  + 如果沒有設定則會自動進行數值索引顯示。
+ 資料類型：每一個 Column 的資料類型，可省略。

### DataFrame 物件
+ DataFrame 是一個表格型的資料結構，每一個 Row 與 Column 都有序號，每一個 Row 與 Column 都可以是不同的資料類型集合 (整數、浮點數、字串)。
+ DataFrame 可以被看作是由Series類似陣列組成的字典集合。
+ DataFrame 中的資料可以用一個或多個二維進行規劃。
+ Numpy.array 中要求所有元素都是相同類型，在 DataFrame 中，但在row 是可以不一樣的。
+ 可用 to_html( ) 方式將物件轉存為網頁，資料將轉為表格。

### 關於索引
+ 預設為數值索引，從 0 開始，亦可以使用 list 方式自建索引。
+ 可以於 pandas 物件建立後，使用 df.index 屬性來更改使用的索引。
+ DataFrame 既有 row 索引也有 column索引。

###  使用Series建立DdataFrame

```python
import pandas as pd
s0 = pd.Series([10,20,30],index=["A","B","C"])
df = pd.DataFrame({"First":s0})
print(df)
data1 = {"Ken":89,"Lucy":61,"Vivin":51}
data2 = {"Ken":78,"Lucy":71,"Vivin":92}
s1 = pd.Series(data1)
s2 = pd.Series(data2)
df = pd.DataFrame({"Score1":s1,"Score2":s2},index = ["Lucy","Ken","Vivin"])
print(df)
print("==========================")
data3 = {"Score1":[72,55,43],"Score2":[92,65,84]}
indexs = ["Ken","Lucy","Vivin"]
df2 = pd.DataFrame(data3,index = indexs)
print(df2)
```

## 資料寫入與讀取
+ 除了存為網頁，也可以存為 csv 與 json 檔案。
+ 寫入檔案：
   + to_csv( ) 方法:第一個參數為欲寫入儲存的檔案，接著可設定索引值是否寫入 (index=False) 與檔案編碼。
   + to_json( ) 方法：寫入到json檔案。
+ 讀取檔案，依據不同的檔案可選擇不同方式開啟：
   + pd.read_html("html檔案")
   + pd.read_csv("csv檔案")
   + pd.read_json("json檔案")

### 關於載入 csv 的參數
+ 打開後預設增加 index 為從 0 自動變化，而 columns 預設使用第一列資料
  + df1 = pd.read_csv('ex1.csv')　
+ header 參數指定 columns 改為從 0 自動變化的整數
  + df2 = pd.read_csv('ex1.csv',header=None)  
+ 使用 names 參數指定 columns 的index
  + df3 = pd.read_csv('ex1.csv',names=list('abcde'))　　　　
+ 使用 index_col 用指定的 columns 的 e欄位作為 row的index值
  + df4 = pd.read_csv('ex1.csv',names=list('abcde'),index_col='e')  

```python
import pandas as pd
import data1
df3 = pd.DataFrame(data1.data1,columns=["name","population"],index=data1.data1["city"])
print(df3)
df3.to_csv("df3.csv",index=False,encoding="utf-8")
df4 = pd.DataFrame(data1.data1)
print(df4)
df4.to_json("df4.json")
```

## 過濾分析
+ 可以透過條件分析的方式針對 row 或指定的 column 進行過濾分析。
+ 可以將過濾的結果轉存為 dataframe。
+ 可以直接於 dataframe 內建立子矩陣方式進行進行過濾檢查。
+ 子矩陣內的操作：
  + 以陣列的概念進行特定欄位的分析，例如 gap1["year"] 欄位名稱中有空白時或特殊字時特別好用 
  + 用.的方式帶出 dataframe 的某特定欄位，例如 gap1.year
+ & 代表兩個條件的 and 邏輯分析，而 | 代表兩個條件的 or 邏輯分析。

```python
import pandas as pd
data_url = 'http://bit.ly/2cLzoxH'
gap1 = pd.read_csv(data_url)
print(gap1.head(3)) #取得前3筆資料
```

```python
import pandas021a
is_2002 =  pandas021a.gap1['year']==2002
# https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.head.html
print(type(is_2002))
print(is_2002.head()) #預設輸出前5筆
print("==================") #預設輸出前5筆
gap1_2002 = pandas021a.gap1[is_2002] # 找year是2002年的
print(gap1_2002.shape)
print(gap1_2002.head())
```

```python
import pandas021a
gap_2002 = pandas021a.gap1[pandas021a.gap1['year']==2002]
print(gap_2002 .shape)
gap2 = pandas021a.gap1[pandas021a.gap1.year == 2002] # 也可使用.year 取代gap1['year'] 指定查找欄位
print(gap2.shape)
gap2 =pandas021a.gap1.query("year == 2002")
print(gap2.shape)
```
